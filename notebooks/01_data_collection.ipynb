```python
import pandas as pd
import numpy as np
import yfinance as yf
import os
import warnings

warnings.filterwarnings("ignore")

pd.set_option("display.max_columns", None)
```

```python
os.makedirs("../data/raw", exist_ok=True)
os.makedirs("../data/processed", exist_ok=True)
```

```python
START = "2015-01-01"
END = "2026-06-30"

tickers = ["TSLA", "SPY", "BND"]

datasets = {}

for ticker in tickers:

    print(f"Downloading {ticker}...")

    df = yf.download(
        ticker,
        start=START,
        end=END,
        auto_adjust=False,
        progress=False
    )

    # Fix MultiIndex columns (new yfinance)
    if isinstance(df.columns, pd.MultiIndex):
        df.columns = df.columns.get_level_values(0)

    df.reset_index(inplace=True)

    datasets[ticker] = df

    df.to_csv(f"../data/raw/{ticker}.csv", index=False)

print("Download Completed")
```

```python
tsla = pd.read_csv("../data/raw/TSLA.csv")
spy = pd.read_csv("../data/raw/SPY.csv")
bnd = pd.read_csv("../data/raw/BND.csv")
```

```python
tsla["Date"] = pd.to_datetime(tsla["Date"])
spy["Date"] = pd.to_datetime(spy["Date"])
bnd["Date"] = pd.to_datetime(bnd["Date"])
```

```python
tsla = tsla.ffill().bfill()
spy = spy.ffill().bfill()
bnd = bnd.ffill().bfill()
```

```python
tsla["Daily Return"] = tsla["Close"].pct_change()

spy["Daily Return"] = spy["Close"].pct_change()

bnd["Daily Return"] = bnd["Close"].pct_change()
```

```python
tsla["Rolling Mean"] = tsla["Close"].rolling(30).mean()
tsla["Rolling Std"] = tsla["Close"].rolling(30).std()

spy["Rolling Mean"] = spy["Close"].rolling(30).mean()
spy["Rolling Std"] = spy["Close"].rolling(30).std()

bnd["Rolling Mean"] = bnd["Close"].rolling(30).mean()
bnd["Rolling Std"] = bnd["Close"].rolling(30).std()
```

```python
tsla.to_csv("../data/processed/TSLA_processed.csv")

spy.to_csv("../data/processed/SPY_processed.csv")

bnd.to_csv("../data/processed/BND_processed.csv")
```
