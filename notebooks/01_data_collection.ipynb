```python
def clean(df):
    df = df.iloc[2:].copy()
    df.rename(columns={"Price": "Date"}, inplace=True)
    df["Date"] = pd.to_datetime(df["Date"])
    df.set_index("Date", inplace=True)
    return df

tsla = clean(tsla)
spy = clean(spy)
bnd = clean(bnd)
```
